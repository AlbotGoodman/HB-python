```python
# Plot med 1:1 dictionary som data

import numpy as np
import matplotlib.pyplot as plt

dic = {'A':1, 'B':3, 'C':2, 'D':4, 'E':3, 'F':5}

k = list(dic.keys())
v = list(dic.values())

plt.plot(k, v)

plt.title('Title')
plt.xlabel('Keys')
plt.ylabel('Values')
plt.show()
```

```python
def Triangle():
    x_values = [0.0]
    y_values = [0.0]
    pos_step = 0
    neg_step = 0
    for x in range(0, 100):
        pos_step += .01
        neg_step -= .01
        x_values.append(pos_step)
        x_values.append(neg_step)
        y_values.append(pos_step)
        y_values.append(neg_step)
    
    x_values.sort()
    y_values.sort()  
    y_values = [abs(x) for x in y_values]
    
    return x_values, y_values

x,y = Triangle()
plt.plot(x, y)
plt.show()
    
plt.plot(Triangle())
plt.show()
```

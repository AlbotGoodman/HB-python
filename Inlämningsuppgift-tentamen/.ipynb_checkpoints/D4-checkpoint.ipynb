### Beskrivning

I denna uppgift skall du skapa en egendefinierad funktion som beräknar prisutvecklingen i 
procentform för de olika kategorier av varor och tjänster samt för olika livsmedel år 1980 – 2021 och 
likt nedan skriver ut och ritar resultatet i en tabell och ett stapeldiagram. Tabellen och diagrammet 
ska baseras på de olika kategorier i de två listorna (tjansteData och livsmedelData). 


```python
# 
# en variant av deluppgift 1 bara för att få fram listorna. 
# 

import csv

def read_file(x):
    with open(x, 'r', encoding='UTF-8') as file:
        data_list = []
        csv_reader = csv.reader(file, delimiter = ';')
        for rad in csv_reader:
            data_list.append(rad)
        return data_list

file1 = 'livsmedel-1.csv'                                     
livsmedelData = read_file(file1)   

file2 = 'tjänster-1.csv'
tjansteData = read_file(file2)   
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# För att slippa de långa namnen
LD = livsmedelData
TD = tjansteData
```

```python
def stapbell():
    
    # data
    
    del LD[0]                                    # Ta bort årtalen
    del TD[0] 

    LDkat = []
    TDkat = []
    for row in LD:                               # Plocka ut kategorierna och lägg i varsin variabel
        LDkat.append(row[0])
        del row[0]
    for row in TD:
        TDkat.append(row[0])
        del row[0]
    
    for a in LD:                                 # Gör om data till floats
        for b, c in enumerate(a):
            a[b] = float(c)
    for a in TD: 
        for b, c in enumerate(a):
            a[b] = float(c)

    LDmax = []                                   # Plocka ut en lista av slutvärdet per rad
    TDmax = []
    for row in LD:
        LDmax.append(row[len(row)-1])
    for row in TD:
        TDmax.append(row[len(row)-1])
    
    LDpro = []                                   # Maxvärdet minus originalvärdet för att få ökningen 
    TDpro = []
    for row in LDmax:
        row = row-100
        LDpro.append(row)
    for row in TDmax:
        row = row-100
        TDpro.append(row)

    
    # tabell
    
    kolumn1_1 = 'Kategorier av varor och tjänster'
    kolumn1_2 = 'Kategorier av olika typer av livsmedel'
    kolumn2 = 'Prisutvecklingen i procentform'
    p = '|'
    outliner = '+-------------------------------------------------------------------------+'
    oneliner = '+-----------------------------------------+-------------------------------+'
    twoliner = '+=========================================+===============================+'
    
    print(outliner)
    print(f'{p}{kolumn1_1:41}{p}{kolumn2}')
    print(twoliner)
    c = 0
    for row in TDkat:
        print(f'{p}{row:41}{p}{TDpro[c]:.2f}')
        print(oneliner)
        c = c + 1
    print(outliner)
    print(f'{p}{kolumn1_2:41}{p}{kolumn2}')
    print(twoliner)
    c = 0
    for row in LDkat:
        print(f'{p}{row:41}{p}{LDpro[c]:.2f}')
        print(oneliner)
        c = c + 1
    
    
    #stapel
    
    plt.figure(figsize=(15,12))
    plt.barh(TDkat, TDpro, color = 'navy')
    plt.barh(LDkat, LDpro, color = 'blue')
    plt.title('Prisutveckling för olika kategorier av varor och tjänster samt för olika typer av livsmedel år 1980-2021 i procentform \n', fontsize = 14)
    plt.xlabel('\n Prisutvecklingen i procentform', fontsize = 14)
    plt.ylabel('Olika kategorier av varor och tjänster samt för olika typer av livsmedel \n', fontsize = 14)
    b = plt.gca()
    b.set_facecolor('whitesmoke')
    plt.show()
```

```python
stapbell()
```

I denna uppgift skall du plotta prisutvecklingen under åren 1980 – 2021 som baseras på innehållet i 
listorna tjansteData eller livsmedelData. Skapa därför en egendefinierad funktion med namnet 
plotta_data som har en lista som inargument. Funktionen ska skapa en av nedanstående diagram 
beroende på om inargumentet är listan tjansteData eller listan livsmedelData.  


```python
# 
# en variant av deluppgift 1 bara för att få fram listorna. 
# 

import csv

def read_file(x):
    with open(x, 'r', encoding='UTF-8') as file:
        data_list = []
        csv_reader = csv.reader(file, delimiter = ';')
        for rad in csv_reader:
            data_list.append(rad)
        return data_list

file1 = 'livsmedel-1.csv'                                     
livsmedelData = read_file(file1)   

file2 = 'tjänster-1.csv'
tjansteData = read_file(file2)   
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

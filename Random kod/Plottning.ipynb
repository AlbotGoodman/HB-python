```python
import matplotlib.pyplot as plt
import numpy as np
import math

#fyller x med hundra tal mellan 0 och 100
x = np.linspace(0, 100, 100)                    #np.linspace(startnummer, stoppnummer, antal)

plt.plot(x, np.sin(2*math.pi*x/100), color = 'red', label = 'sin')
plt.plot(x, np.cos(2*math.pi*x/100), color = 'blue', label = 'cos')

#bakgrundfärg
ax = plt.gca()
ax.set_facecolor('k')

plt.title('Kurvor')
plt.xlabel('X-värde')
plt.ylabel('Y-värde')
plt.legend()

plt.grid()

plt.show()
```
